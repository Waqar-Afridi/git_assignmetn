```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Draw Bar plot
plt.bar([0.25,0.50,0.80,1.0],[20,30,10,40],label='population',color='r',width=0.09)
plt.xlabel('percentage')
plt.ylabel('frequency')
```

```python
# Draw histrogram
```

```python
age=[12,13,32,56,78,90,23,43]
bins=[0,10,20,30,100]
plt.hist(age,bins=bins,histtype='bar',rwidth=0.8)
```

```python
# Scatter plot use for relationship
```

```python
weight=np.array([1,2,3,4,5,6,7,8])
speed=np.array([3,4,5,6,7,8,2,1])
age=np.array([12,23,34,46,78,90,89,76])
plt.scatter(weight,speed)
```

```python
# Area plot
plt.plot([],[],color='g',label=weight,linewidth=5)
plt.plot([],[],color='r',label=speed,linewidth=5)
plt.stackplot(weight,speed,age,color=['c','g'])
```
