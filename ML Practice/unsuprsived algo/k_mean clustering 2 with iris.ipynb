```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
X,y=load_iris(return_X_y=True)
```

kmeans=KMeans(n_clusters=1)

```python
kmeans.fit(X)
```

```python
y_pred=kmeans.predict(X)
```

```python
import matplotlib.pyplot as plt
```
