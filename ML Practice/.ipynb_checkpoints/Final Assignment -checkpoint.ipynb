```python
import nltk
nltk.download('stopwords')
```

```python
import re # regular experision
stemmer = nltk.SnowballStemmer("english") # convert word to its root word
from nltk.corpus import stopwords #common gramitical words that are not use for problem
import string
stopword=set(stopwords.words('english'))
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
```

```python
df['description']=df['description'].apply(clean_text)
```

```python
df.to_csv('final.csv')
```

# Prediction using ML
Now our data is explored and visualized. So its time for prediction using machine learning.

```python
# Features 
X
# Labels
y = df['variety']
```

# Splitting dataset into training and testing
Now, we will split dataset into the training and testing set. Here we have taken 33% data for testing and 77% data for training the model. The code for this is given below:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print('X_train size{} and X_test size{}'.format(X_train.shape,X_test.shape))
```

# Training the model using Naive Bayes
Now in this step we fit our model to the training dataset using Naive Bayes. we are going to create an object of the class named as a nb. The code for this is given below:

```python
# Naive Bayes Classifier
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train,y_train)
```

```python
# Accuracy of our Model
nb = clf.score(X_train,y_train)
print("Accuracy of Naive Bayes: ",nb)
```

# Training the model using logistic regression
Now in this step we fit our model to the training dataset using logistic regression. we are going to create an object of the class named as a lg. The code for this is given below:

```python
# Logistic Regression Classifier
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state = 0)
clf.fit(X_train,y_train)
```

```python
# Accuracy of our Model
lg = clf.score(X_train,y_train)
print("Accuracy of Logistic Regression: ",lg)
```

# Training the model using Support Vector Machine
Now in this step we fit our model to the training dataset using Support Vector Machine. we are going to create an object of the class named as a svm. The code for this is given below:

```python
# SVM Classifier
from sklearn.svm import SVC
clf = SVC(kernel = 'linear', random_state = 0)
clf.fit(X_train,y_train)
```

```python
# Accuracy of our Model
svm = clf.score(X_train,y_train)
print("Accuracy of SVM: ",svm)
```

# Training the model using KNN
Now in this step we fit our model to the training dataset using KNN. we are going to create an object of the class named as a knn. The code for this is given below:

```python
# KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
clf.fit(X_train,y_train)
```

```python
# Accuracy of our Model
knn = clf.score(X_train,y_train)
print("Accuracy of KNN: ",knn)
```

# Training the model using Random Forest
Now in this step we fit our model to the training dataset using Random Forest. we are going to create an object of the class named as a rf. The code for this is given below:

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
```

```python
# Accuracy of our Model
rf = clf.score(X_train,y_train)
print("Accuracy of RF: ",rf)
```

# Training the model using Decision Tree
Now in this step we fit our model to the training dataset using Decision Tree. we are going to create an object of the class named as a dt. The code for this is given below:

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
# Accuracy of our Model
dt = clf.score(X_train,y_train)
print("Accuracy of Decision Tree: ",dt)
```

# Now we will make a comparison graph based on above algorithms predictions.

```python
mylist=[]
mylist2=[]
mylist.append(lg)
mylist2.append("Logistic Regression")
mylist.append(svm)
mylist2.append("SVM")
mylist.append(knn)
mylist2.append("KNN")
mylist.append(nb)
mylist2.append("Naive Bayes")
mylist.append(rf)
mylist2.append("Random Forest")
mylist.append(dt)
mylist2.append("Decision Tree")
plt.rcParams['figure.figsize']=8,6
sns.set_style("darkgrid")
plt.figure(figsize=(18,6))
ax = sns.barplot(x=mylist2, y=mylist, palette = "mako", saturation =1.5)
plt.xlabel("Classification Models", fontsize = 20 )
plt.ylabel("Accuracy", fontsize = 20)
plt.title("Accuracy of different Classification Models", fontsize = 20)
plt.xticks(fontsize = 11, horizontalalignment = 'center', rotation = 8)
plt.yticks(fontsize = 13)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.2%}', (x + width/2, y + height*1.02), ha='center', fontsize = 'x-large')
plt.show()
```
